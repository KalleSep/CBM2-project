PCA is completely unsupervised. Linear discriminant analysis (LDA) is often used for the same purpose as PCA (dimensionality reduction), but is strictly speaking not unsupervised.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
model = LinearDiscriminantAnalysis(n_components = 2, priors = None, shrinkage = 'auto', 
                                   solver = 'eigen', store_covariance = False, tol = 0.0001)
lda = model.fit_transform(X, Y)
plt.scatter(lda[:, 0], lda[:, 1], c = Y, cmap = 'viridis', s = 1)
plt.xlabel("LDA1", fontsize = 20); plt.ylabel("LDA2", fontsize = 20)
feature_importances = pd.DataFrame({'Gene':np.array(expr.columns)[:-1], 
                                    'Score':abs(model.coef_[0])})
print(feature_importances.sort_values('Score', ascending = False).head(20))
```

The way to interpret the data above: we clearly see the data lay in three clusters, suggesting  we have found 3 different separable expression-signatures. However, we also see one cluster is occupied by 2 clusters (the colors are imposed by the fact that we know the "ground truth", but unsupervised methods are generally used for data exploration in which we do not know of these things.

# Non-linear methods

# t-SNE

t-SNE is a very popular decomposition technique used in molecular biology, especially for visualization purposes. t-SNE does generally not cope well with high dimensionality, so it is common to first transform the data with PCA and then run this through t-SNE. Here we will do both with and without prereducing the dimensionality.

```python
from sklearn.manifold import TSNE
```

```python
model = TSNE(learning_rate = 10, n_components = 2, random_state = 123, perplexity = 30)
tsne = model.fit_transform(X)
```

```python
plt.scatter(tsne[:, 0], tsne[:, 1], c = Y, cmap = 'rainbow', s = 2, marker='x')
plt.title('tSNE', fontsize = 20)
plt.xlabel("tSNE1", fontsize = 20)
plt.ylabel("tSNE2", fontsize = 20)
```

```python
#Try different learning rates. How does this change the visualization? Why?
import time

i, j = 0,0
fig, axs = plt.subplots(2,2, figsize=(15,7))

for lr in range(5,25, 5):
    #print(lr)
    start = time.clock()
    model = TSNE(learning_rate = lr, n_components = 2, random_state = 123, perplexity = 30)
    tsne = model.fit_transform(X)
    end = time.clock()    
    axs[i][j].scatter(tsne[:, 0], tsne[:, 1], c = Y, cmap = 'rainbow', s = 2, marker='x')
    axs[i][j].set_title('learning rate{}, time to converge{:.2f}'.format(lr, end-start), fontsize = 10)
    axs[i][j].set_xlabel("tSNE1", fontsize = 5)
    axs[i][j].set_ylabel("tSNE2", fontsize = 5)
    j+=1
    if j%2==0:
        i+=1
        j=0
plt.show()    
    
```

The learning rate controls the step size of the gradiant. The number of iterations to converge vary from dataset to dataset. It is trivial to interpret the results based on hyperparameters such as learning rate, and perplexity.

```python
# effect of perplexity, learning rate and number of iterations in shape 
from itertools import product

parameters = dict(lr= [0, 2, 5, 10, 15, 20, 30],
                 perplexity = [2,5,30, 50, 100],
                 n_iter=[100, 500, 1000, 2000, 5000, 10000])

param_values = [v for v in parameters.values()]


fig = plt.figure(figsize=(25,20))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i, (lr, perplexity, n_iter) in enumerate(product(*param_values)):
    ax = fig.add_subplot(10, 5, i+1)
    start = time.clock()
    model = TSNE(learning_rate = lr, n_components = 2, random_state = 123, perplexity = 30)
    tsne = model.fit_transform(X)
    end = time.clock()    
    ax.scatter(tsne[:, 0], tsne[:, 1], c = Y, cmap = 'rainbow', s = 2, marker='x')
    ax.set_title('lr= {}, perplexity= {} \n n_iter={}, time:{:.2f}'.format(lr, perplexity, n_iter,end-start))
    ax.set_xlabel('t-SNE1')
    ax.set_ylabel('t-SNE2')
    ax.set_axis_off()
plt.show() 
    
    
   
```

```python
from sklearn.cluster import KMeans

# code adapted from https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html#sphx-glr-auto-examples-cluster-plot-kmeans-digits-py

reduced_data = PCA(n_components=2).fit_transform(X)
kmeans = KMeans(init="k-means++", n_clusters=4, n_init=4)
kmeans.fit(reduced_data)


# Step size of the mesh. Decrease to increase the quality of the VQ.
h = 0.02  # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(
    Z,
    interpolation="nearest",
    extent=(xx.min(), xx.max(), yy.min(), yy.max()),
    cmap=plt.cm.Paired,
    aspect="auto",
    origin="lower",
)

plt.plot(reduced_data[:, 0], reduced_data[:, 1], "k.", markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    marker="x",
    s=169,
    linewidths=3,
    color="w",
    zorder=10,
)
plt.title(
    "K-means clustering on the breast cancer dataset (PCA-reduced data)\n"
    "Centroids are marked with white cross"
)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. 

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize


# k-means 
kmeans = KMeans(n_clusters=4).fit(X)

# K-means on normalized data
X_normalized = normalize(X)
Kmeans_normalized = KMeans(n_clusters=4).fit(X_normalized)

# k-means on PCA reduced data
pca_reduced = PCA(n_components=2).fit_transform(X)
kmeans_pca_reduced = KMeans(n_clusters=4).fit(pca_reduced)


print('kmeans: {}'.format(silhouette_score(X, kmeans.labels_, 
                                           metric='euclidean')))
print('kmeans normalized: {}'.format(silhouette_score(X_normalized, Kmeans_normalized.labels_, 
                                           metric='euclidean')))
print('kmeans PCA reduced: {}'.format(silhouette_score(pca_reduced, kmeans_pca_reduced.labels_, 
                                           metric='euclidean')))
```
